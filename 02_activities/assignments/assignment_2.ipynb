  You are given a list containing `n` integers in the range `[0, n]`. Return a list of numbers that are missing from the range `[0, n]` of the array. If there is no missing number, return -1. Note, all the integers in the list may not be unique.
  
  ## Examples

  ### Example 1

  Input: `lst = [0, 2]`

  Output: [1]

  ### Example 2

  Input: `lst = [5, 0, 1]`

  Output: [2, 3, 4]

  ### Example 3

  Input: `lst = [6, 8, 2, 3, 5, 7, 0, 1, 10]`

  Output: [4, 9]

</details>

#### Starter Code for Question 3


```python
from typing import List

def missing_num(nums: List) -> list[int]:
  if not nums:
    return "Warning: Empty List" #flag if the list is empty
  n = max(nums) #detect the maximum number in the list which might be out of order
  expected_set = set(range(n + 1)) #create a set of expected numbers from 0 to n
  observed_set = set(nums)
  missing_numbers = sorted(expected_set - observed_set) #find the difference between expected and observed sets
  return missing_numbers if missing_numbers else [-1] 
```

```python
## Example usage of the missing_num function
# Create named example lists based on the problem statement
example1 = [0, 2] #[1]
example2 = [3, 0, 1] #[2]
example3 = [6, 8, 2, 3, 5, 7, 0, 1, 10] #[4, 9]
# Additional examples 4 through 6 to illustrate additional cases not covered in the problem statement
example4 = [0, 4, 1, 3, 2] #no missing integers = -1
example5 = [5, 6, 8] #integers are missing between 0 and the integers contained in the list = [0, 1, 2, 3, 4, 7]
example6 = [1, 1, 2, 2, 4, 5, 5] #replicates of several integers and missing 0 & 3 = [0, 3]

list_of_examples = [example1, example2, example3, example4, example5, example6]

for i in list_of_examples:
    print("Input:", i, "; Output:", missing_num(i))
```


-   Copy the solution your partner wrote. 


```python
#credit to suchi-dev-ai
def is_valid_brackets(s: str) -> bool:
    # define brackets
    brackets = {')':'(', ']':'[','}':'{'}
    stack = []
    #loop through input
    for char in s:
        #check for closing bracket in input
        if char in brackets:
           #check if stack is empty or top element doesn't match
           if not stack or stack[-1] != brackets[char] :
              return False
           stack.pop()
        else:
            #push bracket into stack
            stack.append(char)
            #return true if stack is empty else false
    return len(stack) == 0
```


-   Explain why their solution works in your own words.


```python
#Their solution works by creating 3 conditions that can return False:
#1. if a closing bracket is encountered and the stack is empty (string starts with a closing bracket)
#2. if a closing bracket is encountered and the top element of the stack does not match the corresponding opening bracket (brackets are not properly nested/ordered)
#3. if there are any unmatched opening brackets left in the stack at the end of the iteration (open brackets without matching closing brackets)

#A "True" return is only possible if the stack is empty at the end of the iteration, meaning all opening brackets had matching closing brackets in the correct order.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# Time Complexity = O(n)
##The string is traversed only once with a for loop, making this approach linear in time complexity
# Space Complexity = O(n)
##The bracket dictionary is constant. The stack stores the opening brackets, which in the worst case is equal to the length of the string, leading to linear space complexity as well. 
```
